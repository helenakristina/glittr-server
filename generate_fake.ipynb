```python
from mimesis import Person, Address
from random import randint
import arrow

current_id, current_artist_id = 0, 0

earliest, latest = 8 * 365, 10 * 365

get_birth_date_offset = lambda: randint(earliest, latest) 

artist = Person("en")
children = []
parents = []
artists = []
parent_children = []
address = Address("en")
for _ in range(100):
    current_id += 1
    current_artist_id += 2
    child_first_name, child_last_name = artist.full_name().split(" ")
    zip_code = address.zip_code()

    now = arrow.utcnow().format()
    child = {
        "artist_id": current_artist_id + 1,
        "child_id": current_id,
        "date_of_birth": arrow.utcnow().shift(days=-get_birth_date_offset()).format(),
        "inserted_dt": now,
        "updated_dt": now
    }
    child_artist = {
        "artist_id": current_artist_id + 1,
        "first_name": child_first_name,
        "last_name": child_last_name,
        "email": artist.email(),
        "password_hash": artist.password(hashed=True),
        "zip_code": zip_code,
        "inserted_dt": now,
        "updated_dt": now
    }
    parent = {
        "artist_id": current_artist_id,
        "parent_id": current_id,
        "inserted_dt": now,
        "updated_dt": now
    }
    parent_artist = {
        "artist_id": current_artist_id,
        "first_name":  artist.full_name().split(" ")[0],
        "last_name": child_last_name,
        "email": artist.email(),
        "password_hash": artist.password(hashed=True),
        "zip_code": zip_code,
        "inserted_dt": now,
        "updated_dt": now
    }
#     parent_child ={
#         "parent_child_id": current_id,
#         "child_id": current_id,
#         "parent_id": current_id,
#         "inserted_dt": now,
#         "updated_dt": now
       
#     }

    children.append(child)
    parents.append(parent)
    artists.append(child_artist)
    artists.append(parent_artist)
    parent_children.append(parent_child)
    
print(children)
print(parents)
print(artists)
```
